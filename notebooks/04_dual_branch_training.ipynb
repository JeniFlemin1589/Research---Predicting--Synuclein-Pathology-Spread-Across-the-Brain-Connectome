# 04 Dual-Branch Training

Train the full GATv2 + GRU RIP model and inspect metrics.

```python
from pathlib import Path
import subprocess
import sys

cwd = Path.cwd()
project_root = cwd.parent if cwd.name == 'notebooks' else cwd
print('Kernel Python:', sys.executable)
print('Project root:', project_root)

subprocess.run(
    [
        sys.executable,
        '-m',
        'ripgnn.cli',
        'train',
        '--config',
        str(project_root / 'configs' / 'demo_dual_branch.yaml'),
    ],
    check=True,
    cwd=project_root,
)
```

```python
subprocess.run(
    [
        sys.executable,
        '-m',
        'ripgnn.cli',
        'interpret',
        '--run-dir',
        str(project_root / 'outputs' / 'demo_dual_branch'),
    ],
    check=True,
    cwd=project_root,
)
```
